```python
img = cv2.imread('/content/drive/MyDrive/Task1 image.jpg')
cv2_imshow(img)
```

```python
gray=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
edges= cv2.Canny(gray, 50, 200)
contours, hierarchy = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
```

```python
areas_list = []
for i in contours:
  area = cv2.contourArea(i)
  areas_list.append(area)
areas_list
```

```python
x = sorted(contours, key=cv2.contourArea, reverse=True)
largest = x[0]
cv2.drawContours(img, largest, -1, (255,255,0), 10)
```

**TASK 2 - SYNOPSIS**

**Computer Vision**

```python
original_image = cv2.imread('/content/drive/MyDrive/Task2 open.jpg')
image = cv2.imread('/content/drive/MyDrive/Task2 open.jpg')
cv2_imshow(original_image)
```

```python
eye_cascade = cv2.CascadeClassifier('haarcascade_eye.xml')
eyes = eye_cascade.detectMultiScale(image, scaleFactor = 1.4, minNeighbors = 5)
for (ex,ey,ew,eh) in eyes:
  cv2.rectangle(image, (ex,ey), (ex+ew,ey+eh), (255,255,255), 3)
cv2_imshow(image)
```
